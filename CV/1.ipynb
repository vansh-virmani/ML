```python
import cv2
```

```python
##cv2 is the standard Library alias used to OpenCV in Python 
## we get all the major image processing functions from cv2
```

```python
img=cv2.imread('C:/Users/acer/Downloads/rs.jpg')
```

```python
cv2.imshow("rs",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Convert image into grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2.imshow('GRAY rs', gray)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
