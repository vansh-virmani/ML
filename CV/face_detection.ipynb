```python
import cv2
```

```python
haar_features=cv2.CascadeClassifier('haarcascade_frontalface_alt.xml')
```

```python
cap=cv2.VideoCapture(0)
while True:
    flag,frame=cap.read()
    if not flag:
        print("Camera not available")
        break
    faces=haar_features.detectMultiScale(frame,scaleFactor=1.3)
    # faces=haar_features.detectMultiScale(img)
    for (x,y,w,h) in faces:
      cv2.rectangle(frame,(x,y),(x+w,y+h),(0,255,0),2)
    cv2.imshow('output',frame)
    key=cv2.waitKey(25) & 0xFF
    if key==27:
      break
cap.release()
cv2.destroyAllWindows()
```
