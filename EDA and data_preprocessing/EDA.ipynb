```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#cleaning and feature engineering process:
df['Date']=df['Date_of_Journey'].str.split('/').str[0]
df['Month']=df['Date_of_Journey'].str.split('/').str[1]
df['Year']=df['Date_of_Journey'].str.split('/').str[2]
#broken down into parts as a new feature but now we have to convert it into
# int as it is still object categorical variable.
```

```python
df['Date']=df['Date'].astype(int)
df['Month']=df['Month'].astype(int)
df['Year']=df['Year'].astype(int)
```

```python
df.info() #converted to int 
```

```python
df.drop('Date_of_Journey',axis=1,inplace=True)#axis=1 remove columns 
#axis=0 drop rows and inplace=true directly make changes in og dataframe
```

```python
#similarly for arrival time
df['arrival_hour']=df['Arrival_Time'].str.split(':').str[0]
df['arrival_min']=df['Arrival_Time'].str.split(' ').str[0]
df['arrival_min']=df['arrival_min'].str.split(':').str[1]
```

```python
df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
df['arrival_hour']=df['arrival_hour'].astype(int)
df['arrival_min']=df['arrival_min'].astype(int)
```

```python
df.info()
```

```python
#Similarly for Departure time
df['dep_hour']=df['Dep_Time'].str.split(':').str[0]
df['dep_min']=df['Dep_Time'].str.split(':').str[0]
```

```python
df['dep_hour']=df['dep_hour'].astype(int)
df['dep_min']=df['dep_min'].astype(int)
```

```python
df.head()
df.info()
```

```python
df.drop('Dep_Time',axis=1,inplace=True)
```

```python
#more stops more price so give some labels to it of flight
#map or do in each line by using replace
df['Total_Stops']=df['Total_Stops'].map({'non-stop':0, '2 stops':2, '1 stop':1, '3 stops':3,  '4 stops':4,np.nan:1})
```

```python
#drop route as we have source and dest
df.drop('Route',axis=1,inplace=True)
```

```python
df['Duration_hour'] = df['Duration'].str.split(' ').str[0] \
                                     .str.split('h').str[0] \
                                     .where(df['Duration'].str.contains('h'), 0) \
                                     .astype(int)
#.where is as if else inline saying if there is no h keep 0
df['Duration_hour'].unique()
#here also duration hour contains 5m which will cause error so handle this
```

```python
df['Duration_min']=df['Duration'].str.split(' ').str[1].str.split('m').str[0]
df['Duration_min']=df['Duration_min'].replace(np.nan,0)
```

```python
df['Duration_min']=df['Duration_min'].astype(int)
```
